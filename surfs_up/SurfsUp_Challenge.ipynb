```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import pandas as pd

import datetime as dt
```

```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 

# Calculate the date one year from the last date in data set.
# Use the filter() function to filter out the data not needed.
# Add a function that extracts 'all' of the results from the query and put them in a list. 
#print(results)

june_temps = session.query(Measurement).filter(extract('month', Measurement.date) == 6)
june_temps = session.query(Measurement).filter(extract('month', Measurement.date) == 6).all()

print(june_temps)
```

```python
# 3. Convert the June temperatures to a list.

# june_temps = [] -- 'for' list

june_temps = [temp.tobs for temp in june_temps]
june_temps
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 

# Convert the results to a DataFrame
#df = pd.DataFrame(results, columns=['tobs'])

june_temp_df = pd.DataFrame(june_temps, columns=['June Temps'])
june_temp_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.

# Use Pandas to calcualte the summary statistics for the precipitation data
# df.describe()

# Use Pandas to calculate the summary statistics for the precipitation data
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.

dec_temps = session.query(Measurement).filter(extract('month', Measurement.date) == 12)
dec_temps = session.query(Measurement).filter(extract('month', Measurement.date) == 12).all()

print(dec_temps)
```

```python
# 7. Convert the December temperatures to a list.

dec_temps = [temp.tobs for temp in dec_temps]
dec_temps
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 

dec_temp_df = pd.DataFrame(dec_temps, columns=['December Temps'])
dec_temp_df
```
